```python
bin_age = [0, 19, 29, 39, 49, 59, 69, 80]
category_age = ['<20s', '20s', '30s', '40s', '50s', '60s', '>60s']
df['Age_binned'] = pd.cut(df['Age'], bins=bin_age, labels=category_age)
df = df.drop(['Age'], axis = 1)
```

```python
bin_NatoK = [0, 9, 19, 29, 50]
category_NatoK = ['<10', '10-20', '20-30', '>30']
df['Na_to_K_binned'] = pd.cut(df['Na_to_K'], bins=bin_NatoK, labels=category_NatoK)
df = df.drop(['Na_to_K'], axis = 1)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(["Drug"], axis=1)
y = df["Drug"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)

```

```python
cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test,y_pred)
acc_score = accuracy_score(y_test,y_pred)
print(cm)
print(cr)
print(acc_score*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
R_model = RandomForestClassifier(max_leaf_nodes=20)
R_model.fit(X_train,y_train)
```

```python
y_pred = R_model.predict(X_test)
```

```python
cm_r = confusion_matrix(y_test,y_pred)
cr_r= classification_report(y_test,y_pred)
acc_score_r = accuracy_score(y_test,y_pred)
print(cm_r)
print(cr_r)
print(f"{acc_score_r*100}%")
```
